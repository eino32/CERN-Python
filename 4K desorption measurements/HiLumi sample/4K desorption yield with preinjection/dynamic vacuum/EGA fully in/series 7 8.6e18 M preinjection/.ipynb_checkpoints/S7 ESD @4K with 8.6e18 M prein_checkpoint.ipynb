### Using linear regression, remove the pressure decrease from pumping

```python
# Load the data
t = cut_time["Elapsed time, mass 2"]
print(t.describe())
I_ion = cut_hv["Ion current"]
print(I_ion.describe())
#The function for the ion current-time relationship is not known. 
#numpy method for creating a polynomial model
model = np.poly1d(np.polyfit(t, I_ion, 8))

# Print the original function
plt.figure(figsize=(12,6))
plt.plot(t, I_ion)

# Add labels and title
plt.xlabel('time (s)')
plt.ylabel('Ion current')
plt.title('Ion current vs time')

plt.yscale("log")
plt.show()
```

```python
#testing and plotting

# Define the range of positions to plot
x = np.linspace(5095.951000, 19700, 3095)


# Plot the function
plt.figure(figsize=(12,6))
plt.plot(x, model(x))

# Add labels and title
plt.xlabel('time (s)')
plt.ylabel('Ion current')
plt.title('Ion current vs time')

plt.yscale("log")
plt.show()
```

```python
#subtract fitted data from real data
I_norm = I_ion/model(x)

# Plot the normalised ion current
plt.figure(figsize=(12,6))
plt.plot(t, I_norm)

# Add labels and title
plt.xlabel('time (s)')
plt.ylabel('normalised ion current (a.u.)')
plt.title('Ion current vs time')

plt.yscale("linear")
plt.show()
```

```python
#find the middle of the desorption peaks
esd = cut_hv[cut_hv["Elapsed time"].between(10370,19530)]

#find the average of base ion current
base_I_ion = esd["Ion current"].mean()
print(base_I_ion)

#multiply normalised ion current by average of base ion current
I_ion_corr = I_norm*base_I_ion
print(I_ion_corr)

# Plot the result
plt.figure(figsize=(12,6))
plt.plot(t, I_ion_corr)

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Corrected H2 ion current (A)')
plt.title('Ion current vs time')

plt.yscale("linear")
plt.show()
```

```python
#add the new values to cut_hv dataframe
cut_hv.insert(23, "Corrected ion current", I_ion_corr)
```

### Conversion from H2 ion current to partial pressure:

```python
#Filtering background signal for resampled data

#take a mass with a signal known to be in the background as a definition of electronic noise
#Ignored because has negligeable impact (~1%)
#elec_bck = np.mean(mid["mass 5.5_value"])

# take the Ion current of H2 channel before emission start as a reference to base pressure:
#get the rows where there was no emission, before electron activity
res1 = cut_hv.loc[cut_hv["Comment"]=="electron activity"]
res2 = cut_hv.loc[cut_hv["Comment"]=="grid transparent"]

pre_EM = cut_hv.loc[cut_hv["Elapsed time"] <= res1["Elapsed time"][0]]

print(pre_EM.describe())
```

```python
dyn_bck = np.quantile(pre_EM["Corrected ion current"],0.25)
print(dyn_bck)

#apply background corrections 
#NB! replace hv_aligned["Ion current"] with 
cut_hv["mass 2 corr"] = cut_hv["Corrected ion current"] - dyn_bck #- static bck
#remove values below 0:
cut_hv["mass 2 corr"].clip(0, inplace=True)
print(cut_hv.head())

# Plot the result
plt.figure(figsize=(12,6))
plt.plot(t, cut_hv["mass 2 corr"])

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('Background subtracted H2 ion current (A)')
plt.title('H2 Ion current vs time')
plt.xlim(10000,20000)
plt.yscale("linear")
plt.show()
```

```python
#From the signal, get the partial pressures 
h2_sens = 2.13e-3 #[A/mbar] approximation for faraday mode
p_h2_ip = cut_hv["mass 2 corr"]/(h2_sens*0.46) #H2 equivalent pressures


#account for the thermal transpiration factor (electrons thermalising to RT)
T2 = 4.2
T1 = 295.5
T_coef = np.sqrt(T2/T1)
p_h2_ip = p_h2_ip*T_coef 
```

```python
#insert partial pressures into the dataframe
cut_hv.insert(22,"hydrogen pp",p_h2_ip)
cut_hv.fillna(0,inplace=True)

cut_hv["hydrogen pp"].describe()
```

## H2 Desorption yield point-by-point

$$
\eta=\frac{C_{H2}\cdot (p_{n,EM} - \overline{p}_{base}) \cdot q_e}{k_B\cdot T \cdot I_{n}}
$$


```python
#partial pressure delta
#Wrong methodology below. Need to subtract p_duringEM form p_base, noEM
#p_delta = hv_aligned["hydrogen pp"].diff(periods=1).fillna(hv_aligned["hydrogen pp"])*100 #conversion to Pa
#p_delta = p_delta.clip(lower=0)

#Correct methodology
#for base pressure, use pre_EM datapoints
pre_EM_corr = cut_hv.loc[cut_hv["Elapsed time"] <= res1["Elapsed time"][0]]

mean_pbase = np.mean(pre_EM_corr["hydrogen pp"])
print(mean_pbase)

EM = cut_hv.loc[cut_hv["Elapsed time"] >= res2["Elapsed time"][0]]
p_delta = (EM["hydrogen pp"] - mean_pbase)*100 #mbar to Pa
p_delta.clip(0, inplace=True)
print(p_delta)
```

```python
#constants are already defined
#emission current and H2 partial pressures are now available at the same sampling rates
I_n = EM["I_em"].replace(0, np.nan)

#pd.df.where(cond= "Where cond is True, keep the original value. Otherwise, replace with NaN")
I_n.where(EM["I_em"] >= 9e-7, np.nan, inplace=True) #--- this will cause minor misalignment if NaNs are thrown away
I_n.dropna(inplace=True)
print(I_n.describe())

eta = ((x_h2*p_delta*q_e)/(k_B*T*I_n)).dropna()
#ignore first row to match "i" dataset
eta = eta.loc[eta.index[1]:]

#eta = eta
print("desorption yields point-by-point:","\n", eta)
```

```python
#slicing eta to match hv_aligned["Elapsed time"]:
# Slice x using the index of y
hv_downsampled = hv_aligned.loc[eta.index, :]
hv_downsampled.describe()
```

## Electron dose


Find:

$$
Q = \int_{t_1}^{t_2} I \,dt
$$

Where $t_2$ and $t_1$ are the timestamps at which an emission current reading was taken


```python
#Integrating the emission current

def integrand(t, i):
    return i*t

i = I_n.dropna().reset_index() 

#for testing:
#t2_2 = (hv_aligned["Elapsed time"].diff(periods=1).fillna(0))

#adding an elapsed time column
timestamp3=[]
for j in hv["Time"]:
    timestamp3.append(j)
    
#making the column into a datetime object
timestamps3=pd.to_datetime(i["datetime"],format="%H:%M:%S")
runtime3=(timestamps3-timestamps3[0]).dt.total_seconds()    
i.insert(2,"Elapsed time",runtime3)

#define the n, n-1 difference in a variable "t2"
t2 = (i["Elapsed time"].diff(periods=1).fillna(0))

#replace all values from t2 that are further away than 50% of the median (.. > 9) with the median value
t2.where(t2 <= 12,6,inplace=True)
print(t2)

#integration
ar = []
for x in t2:
    itr = 0
    I = quad(integrand, 0, x, args=(i["I_em"][itr]))
    ar.append(I[0])
    itr += 1
    #if previous value was not NaN, replace the value with the median of elapsed time
    #if 
    
```

```python
#calculate the cumulative charge
Q_cumul = np.cumsum(ar)
Q_sum = sum(ar)
print("Total charge emitted from EGA during ~5000 seconds:", f"{Q_sum:.4E}", "C")

#cumulative nr of electrons emitted
N_e = Q_sum/1.6e-19
Ne_cumul = Q_cumul/1.6e-19
print("Total number of electron emitted during ~5000 seconds:", f"{N_e:.4E}", "e")
      
#To get the electron dose, divide n of electrons by sample surface area of 276 cm2, multiply by grid transmittance of 0.9
tot_dose = (N_e/276)*0.9
cumul_dose = (Ne_cumul/276)*0.9
print("Total electron dose on the sample ~5000 seconds:", f"{tot_dose:.4E}", "e-/cm^2")
      
#add these to the "i" dataframe:
i.insert(3,"Cumulative charge", Q_cumul)
i.insert(4,"Cumulative n. of electrons emitted from EGA", Ne_cumul)
i.insert(5,"Cumulative electron dose on the HL sample", cumul_dose)
i.set_index("datetime", drop=True, inplace=True)

#Ignore the first value, also, match the shape of eta
i = i.loc[i.index[1]:]
print(i)
```
