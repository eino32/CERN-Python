# Desorption yield for H2, H2O, CO, CO2
The desorption yield is given by formula:

$$
\eta=\frac{Q\cdot q_e}{k_B\cdot T \cdot I}=\frac{\Delta p \cdot S_{eff}\cdot q_e}{k_B\cdot T \cdot I_{Em}}
$$

For the RT setup, we use flux units of $[\frac{mbar\cdot l}{s}]$ and substitute pressure change over time multiplied by chamber volume $[\frac{\Delta p\cdot V}{s}]$ with pressure change $[\Delta p]$ and effective pumping speed $[\frac{l}{s}]$.

## Procedure
1) Calculate the difference in pressure before and after turning emission on <br/>
2) Take the effective pumping speed value from the JUAS protocol (calculated for air) <br/>
3) Assume for now that the temperature is constant at 23 degrees = 296 K <br/>
4) Assume for simplicity that H2 is the dominating gas whose relative quantity in the system is 1 <br/>
5) $x=I_{H2}/I_{ion}$ <br/>
6) $x\cdot I_{Em}$=weighted ion current for a gas  <br/>
7) Calculate the desorption yield for each species and plot vs quantity of desorbed gas ($\Delta p\cdot S_{eff}$) <br/>
NB! Need to subtract the EGA emission from the total signal

```python
#read in the pressure file
p= pd.read_csv("processed data, grid transparent", sep=',')
p.drop(columns="Unnamed: 0")
#Use penning 2 readings, more representative of the part of the chamber with electron activity
```

```python
#step 1: H2 min and max pressures unit: Pa
delta_p=(p["Penning_2"].max()-p["Penning_2"].min())*100
print("Pressure delta in Pa:",delta_p)
#step 2 unit: m3/s
S_eff=42/1000
#step 5
x_1=1 #for H2
x_2=df["Ch2"].max()/df["Ch1"].max()
x_3=df["Ch3"].max()/df["Ch1"].max()
x_4=df["Ch4"].max()/df["Ch1"].max()
x_5=df["Ch5"].max()/df["Ch1"].max()
x_6=df["Ch6"].max()/df["Ch1"].max()
x_7=df["Ch7"].max()/df["Ch1"].max()
x_8=df["Ch8"].max()/df["Ch1"].max()
x_9=df["Ch9"].max()/df["Ch1"].max()
x_10=df["Ch10"].max()/df["Ch1"].max()
print(x_2,x_3,x_4,x_5,x_6)

#step 6 multiply maximum EGA emission current by weight
w_1=x_1*p["I_Em"].max()
w_2=x_2*p["I_Em"].max()
w_3=x_3*p["I_Em"].max()
w_4=x_4*p["I_Em"].max()
w_5=x_5*p["I_Em"].max()
w_6=x_6*p["I_Em"].max()
w_7=x_7*p["I_Em"].max()
w_8=x_8*p["I_Em"].max()
w_9=x_9*p["I_Em"].max()
w_10=x_10*p["I_Em"].max()
print("weighted ion current for C", w_2)
## Define constants
k_B=1.38e-23
T=296
q_e=1.6e-19
#step 7
eta_1=(delta_p*S_eff*q_e)/(k_B*T*w_1)
eta_2=(delta_p*S_eff*q_e)/(k_B*T*w_2)
eta_3=(delta_p*S_eff*q_e)/(k_B*T*w_3)
eta_4=(delta_p*S_eff*q_e)/(k_B*T*w_4)
eta_5=(delta_p*S_eff*q_e)/(k_B*T*w_5)
eta_6=(delta_p*S_eff*q_e)/(k_B*T*w_6)
eta_7=(delta_p*S_eff*q_e)/(k_B*T*w_7)
eta_8=(delta_p*S_eff*q_e)/(k_B*T*w_8)
eta_9=(delta_p*S_eff*q_e)/(k_B*T*w_9)
eta_10=(delta_p*S_eff*q_e)/(k_B*T*w_10)
eta_ar=[eta_1,eta_2,eta_3,eta_4,eta_5,eta_6,eta_7,eta_8,eta_9,eta_10]
#fluxes under EGA bombardment
cols=["Ch1","Ch2","Ch3","Ch4","Ch5","Ch6","Ch7","Ch8","Ch9","Ch10"]
I_sum=df[cols].sum(axis=0)
print(I_sum)
dp=[]
for i in cols:
    dp.append(df[i].max()/I_sum[i])

#praegu jagab iga kanali maksimumväärtuse läbi 10 erineva I_sumi väärtusega
fluxes=[element * S_eff for element in dp]
```
