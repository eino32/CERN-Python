```python
#read in the data
mid = pd.read_csv("MID.tsv", sep="\t", skiprows=lambda x: x <=11,  on_bad_lines="skip")
mid.drop([0,1], inplace=True)
#print(mid.describe())
#bring the time data into a new df:
time = pd.DataFrame(mid.iloc[:,0::2]) #- get every second column
time["mass 2.0_time"] = pd.to_datetime(time["mass 2.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 4.0_time"] = pd.to_datetime(time["mass 4.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 5.5_time"] = pd.to_datetime(time["mass 5.5_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 12.0_time"] = pd.to_datetime(time["mass 12.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 15.0_time"] = pd.to_datetime(time["mass 15.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 16.0_time"] = pd.to_datetime(time["mass 16.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 18.0_time"] = pd.to_datetime(time["mass 18.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 28.0_time"] = pd.to_datetime(time["mass 28.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 32.0_time"] = pd.to_datetime(time["mass 32.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 40.0_time"] = pd.to_datetime(time["mass 40.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')
time["mass 44.0_time"] = pd.to_datetime(time["mass 44.0_time"], format="%Y/%m/%d %H:%M:%S.%f").dt.strftime('%d/%m/%Y %H:%M:%S')

#add elapsed time columns
def elapsed_time(df):
    itr = 0
    counter = len(df.columns)
    masses = [2,4,5.5,12,15,16,18,28,32,40,44]
    for i in df:
        timestamp = pd.to_datetime(df[i], format="%Y/%m/%d %H:%M:%S.%f")
        runtime = (timestamp-timestamp[2]).dt.total_seconds()
        colname = "Elapsed time, mass {}".format(masses[itr]) 
        df.insert(counter, colname, runtime)
        itr += 1
        counter += 1
    return(df)
#elapsed_time(time)
time
#time["Elapsed time, mass 40"].describe()
```

### Checking injection pressure linearity

```python
#defining a function for the straight line
def func(x,a,b):
    return a*x + b
#curve fit for the data
y = vaclog["injection 1mbar"].values
x = (vaclog["Elapsed time"]/3600).values
params, cov = curve_fit(func,x,y)
#straight line parameters
a, b = params
print(f"Fitted line: y = {a:.2f}*x+{b:.2f}")
```

```python
#checking inj pressure linearity, graph
fig, ax = plt.subplots()
y_fit = func(x,a,b)
ax.plot((vaclog["Elapsed time"]/3600),vaclog["injection 1mbar"],marker=".", markersize=5,label='inj volume pressure')
ax.plot((vaclog["Elapsed time"]/3600),y_fit, "--",color="red", label="Y= -ax + b fit") 
ax.legend(loc="upper right")

ax.set_xlabel('Elapsed time (h)')
ax.set_ylabel('Pressure (mbar)')
ax.set_yscale('linear')
plt.title("S1 Injection pressure evolution 1mbar conductance gauge")
```

```python
#Temperature curve for CERNOX - for temp stability
A=[230.317302,-6170.1513,71837.9529,-477946.76,2.003668910085786e+6,-5.488690193047771e+6,9.830475663897528e+6,-1.111226817786569e+7,7.202477878914065e+6,-2.04194551328507e+6]
#specify fit parameters A, data (Resistance values)
def polyfit(params,data):
    total=[]
    for j in data: 
        exp=0
        for i in range(len(params)):
            exp += (params[i]/(math.log10(j))**i)
        total.append(10**exp)
    return(total)    
```

# Primary desorption calculation 
### Elena's formula for the cryo setup
$$
\eta=\frac{C_{H2}\cdot (\Delta p_1 - \Delta p_2)q_e}{k_B\cdot T \cdot I_e}
$$

Where $\Delta p_i=p(t)_i - p(t=0)_i$ and $i=1,2$
In the case of the 4K setup, we know the exact conductance of gasses. The formula above applies well.

Due to the EGA being fully retracted, it is safe to assume that the desorption takes place at approximately room temperature

```python
#calculation delta_p-s
dp1 = (np.mean(p_h2)-np.mean(base_h2))*100
print(dp1)

#Constants
k_B = 1.38e-23
T = 293
print(T)
q_e = 1.6e-19
c_h2 = 37.91/1000
s_h2=42/1000
x_h2=1/((1/c_h2)+(1/s_h2))
print(x_h2)
#Emission current
#ignore the inrush current values
hv["em"].where(hv["em"] <= 1.5e-6, np.nan, inplace=True)

I_em = np.nanmean(hv["em"].nlargest(30))
print(I_em)
#Conductance for CO in m3/s
c_co=10.13/1000
c_co2 = 8.08/1000
```

```python
#Desorption yield
eta_h2 = (x_h2*dp1*q_e)/(k_B*T*I_em)
#eta_co = (c_co*(dp1-dp2)*q_e)/(k_B*T*I_em)
print("S1 The desorption yield for H2 [mol/el]:",eta_h2,"\n",)#""The desorption yield for CO2 [mol/el]:",eta_co2)
```
